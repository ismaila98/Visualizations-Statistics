Analysis
---

- Statistics show that Infubinol & Rumicane have a Median Tumor Volume that is almost centered within the interquartile range of occupancy, which concludes that values aren't skewed away from each other. Only one outlier appears to be present within the Infubinol group.

- Within the Capomulin group, there appears to be a strong positive correlation between a mice's weight and its tumor volume.

- Capomulin and Rubicane appear to be the best treatments overall. Mouse within these two Drug Regimens appear to have the lowest median tumor volume by the end of the study.

- Analyzing Median Tumor Volume, Capomulin has a smaller interquartile range than that of Rubicane. Thus, Rubicane has a smaller Median Tumor Volume than that of Capomulin.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
len(merge_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
merge_df.loc[merge_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.loc[(merge_df['Mouse ID']!="g989")]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count = len(clean_df["Mouse ID"].unique())
mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_values = clean_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median_values = clean_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
var_values = clean_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
stdev_values = clean_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sem_values = clean_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
# Assemble the resulting series into a single summary DataFrame.
summary_df = pd.DataFrame({ "Mean Tumor Volume": mean_values, "Median Tumor Volume": median_values,
                                   "Tumor Volume Variance": var_values,
                                   "Tumor Volume Std. Dev.": stdev_values,
                                   "Tumor Volume Std. Err.": sem_values,
                                   })
summary_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
pandas_plot = clean_df.groupby('Drug Regimen').count()['Mouse ID']
pandas_plot = pandas_plot.sort_values(ascending=False)
pandas_plot.plot(kind="bar", figsize=(5,3))
plt.title(" Total Number of Mice Tested For Each Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice Tested")

plt.show()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
x_axis = np.arange(len(pandas_plot))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(5,3))
plt.bar(x_axis, pandas_plot, alpha=.75, align='center')
plt.xticks(tick_locations, pandas_plot.index.values, rotation="vertical")

plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(pandas_plot)+15)

plt.title("Total Number of Mice Tested For Each Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice Tested")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
male_female = clean_df.groupby(["Sex"]).count()["Mouse ID"]
labels = ["Female", "Male"]
explode = (0.02,0)
colors = ["orange", "blue"]
plt.pie(male_female, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", startangle=180)
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

greatest_timepoint = clean_df.groupby('Mouse ID').max()['Timepoint']
greatest_timepoint
merging_df = pd.merge(greatest_timepoint, clean_df, on = ("Mouse ID", "Timepoint"), how = "inner" )
merging_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_volume = []

for drug in treatments:
    final_tumor_volume = merging_df.loc[merging_df["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    tumor_volume.append(final_tumor_volume)
    

capomulin = tumor_volume[0]
ca_quartiles = capomulin.quantile([0.25,0.5,0.75])
capomulin_lq = ca_quartiles[0.25]
capomulin_medq = ca_quartiles[.5]
capomulin_uq = ca_quartiles[0.75]
capomulin_iqr = capomulin_uq - capomulin_lq
print('Capomulin')
print(f'The lower quartile of occupancy is: {capomulin_lq}')
print(f'The upper quartile of occupancy is: {capomulin_uq}')
print(f'The interquartile range of occupancy is: {capomulin_iqr}')
print(f'The median of occupancy is: {capomulin_medq}')
capomulin_lowerbound = capomulin_lq - 1.5*capomulin_iqr
capomulin_upperbound = capomulin_uq + 1.5*capomulin_iqr
print(f'Lower Bound: {capomulin_lowerbound}')
print(f'Upper Bound: {capomulin_upperbound}')
print(f'Values below {capomulin_lowerbound} could be outliers.')
print(f'Values above {capomulin_upperbound} could be outliers.')

    
```

```python
ramicane = tumor_volume[1]
r_quartiles = ramicane.quantile([0.25,0.5,0.75])
ramicane_lq = r_quartiles[0.25]
ramicane_medq = r_quartiles[.5]
ramicane_uq = r_quartiles[0.75]
ramicane_iqr = ramicane_uq - ramicane_lq
print('Ramicane')
print(f'The lower quartile of occupancy is: {ramicane_lq}')
print(f'The upper quartile of occupancy is: {ramicane_uq}')
print(f'The interquartile range of occupancy is: {ramicane_iqr}')
print(f'The median of occupancy is: {ramicane_medq}')
ramicane_lowerbound = ramicane_lq - 1.5*ramicane_iqr
ramicane_upperbound = ramicane_uq + 1.5*ramicane_iqr
print(f'Lower Bound: {ramicane_lowerbound}')
print(f'Upper Bound: {ramicane_upperbound}')
print(f'Values below {ramicane_lowerbound} could be outliers.')
print(f'Values above {ramicane_upperbound} could be outliers.')
```

```python
infubinol = tumor_volume[2]
i_quartiles = infubinol.quantile([0.25,0.5,0.75])
infubinol_lq = i_quartiles[0.25]
infubinol_medq = i_quartiles[.5]
infubinol_uq = i_quartiles[0.75]
infubinol_iqr = infubinol_uq - infubinol_lq
print('Infubinol')
print(f'The lower quartile of occupancy is: {infubinol_lq}')
print(f'The upper quartile of occupancy is: {infubinol_uq}')
print(f'The interquartile range of occupancy is: {infubinol_iqr}')
print(f'The median of occupancy is: {infubinol_medq}')
infubinol_lowerbound = infubinol_lq - 1.5*infubinol_iqr
infubinol_upperbound = infubinol_uq + 1.5*infubinol_iqr
print(f'Lower Bound: {infubinol_lowerbound}')
print(f'Upper Bound: {infubinol_upperbound}')
print(f'Values below {infubinol_lowerbound} could be outliers.')
print(f'Values above {infubinol_upperbound} could be outliers.')
```

```python
ceftamin = tumor_volume[3]
ce_quartiles = ceftamin.quantile([0.25,0.5,0.75])
ceftamin_lq = ce_quartiles[0.25]
ceftamin_medq = ce_quartiles[.5]
ceftamin_uq = ce_quartiles[0.75]
ceftamin_iqr = ceftamin_uq - ceftamin_lq
print('Ceftamin')
print(f'The lower quartile of occupancy is: {ceftamin_lq}')
print(f'The upper quartile of occupancy is: {ceftamin_uq}')
print(f'The interquartile range of occupancy is: {ceftamin_iqr}')
print(f'The median of occupancy is: {ceftamin_medq}')
ceftamin_lowerbound = ceftamin_lq - 1.5*ceftamin_iqr
ceftamin_upperbound = ceftamin_uq + 1.5*ceftamin_iqr
print(f'Lower Bound: {ceftamin_lowerbound}')
print(f'Upper Bound: {ceftamin_upperbound}')
print(f'Values below {ceftamin_lowerbound} could be outliers.')
print(f'Values above {ceftamin_upperbound} could be outliers.')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
treatments_plot = [capomulin, ramicane, infubinol, ceftamin]

fig1, ax1 = plt.subplots()
ax1.set_title('Tumor Volume by Treatment')
ax1.set_xlabel('Treatment')
ax1.set_ylabel('Tumor Volume (mm3)')

ax1.boxplot(treatments_plot, labels=["Capomulin","Ramicane","Infubinol","Ceftamin"])
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_mouse = clean_df.loc[(clean_df['Mouse ID'] == 's185')]
capomulin_mouse = capomulin_mouse.set_index('Timepoint')
capomulin_mouse['Tumor Volume (mm3)'].plot(color = 'blue')
plt.title("Mouse s185's Tumor Over Time")
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatterp = clean_df.loc[(clean_df['Drug Regimen'] == 'Capomulin')]
scatterp = scatterp.groupby(["Mouse ID"]).mean()
weight_plot = scatterp["Weight (g)"]
tumor_plot = scatterp["Tumor Volume (mm3)"]
plt.scatter(weight_plot, tumor_plot, color = "blue")
plt.xlabel("Weight (g)")
plt.ylabel("Avg. Tumor Volume (mm3)")
plt.title("Avg. Tumor Volume vs Weight for Capomulin Treatment")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
print(f"The correlation coefficient between average tumor volume and weight is {round(st.pearsonr(tumor_plot,weight_plot)[0],2)}")
```

```python
linear_reg = st.linregress(scatterp['Tumor Volume (mm3)'], scatterp['Weight (g)'])
linear_reg
```

```python
x = clean_df['Weight (g)']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(weight_plot, tumor_plot)
regress_values = x * slope + intercept
plt.scatter(weight_plot, tumor_plot, color = "red")
plt.plot(x, regress_values, color = "yellow")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Linear Regression for Capomulin Treatment")
plt.xlim(14.5, 26)
plt.ylim(34, 46.5)
plt.show()
```
